```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
os.getcwd()
```

```python
file_name = 'Analyst_case_study_dataset.xlsx'
sheet_name = 'Sheet1'
```

```python
df = pd.read_excel(file_name, sheet_name=sheet_name)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
nulls = df.isnull().sum().reset_index().rename(columns={'index':'columns', 0:'count_nulls'})
nulls['null_ratio'] = (nulls['count_nulls']/len(df)) * 100
nulls
```

```python
# All null values in IP Address, so we remove the entire columns
df.drop(columns='IP Address', inplace=True)
```

```python
df.info()
```

```python
df['CallStatus'] = df['CallStatus'].fillna('Unknown')
```

```python
df['CallStatus'].value_counts()
```

```python
df['LeadCategory'] = df['CallStatus'].map({"Closed":"Good", "Contacted - Doesn't Qualify":"Bad", "Unable to contact - Bad Contact Information":"Bad", "EP Confirmed":"Good", "Contacted - Invalid Profile":"Bad", "EP Sent":"Good", "EP Received":"Good", "Unknown":"Unknown"})
```

```python
df['LeadCategory'].value_counts()
```

```python
# % of goof leads generated'
len(df[df['LeadCategory'] == 'Good'])/len(df) * 100
```

```python
# % of good leads generated out of known ones'
len(df[df['LeadCategory'] == 'Good'])/len(df[df['LeadCategory'] != 'Unknown']) * 100
```

```python
df['LeadClosedFlag'] = df['CallStatus'].map({"Closed":1, "Unknown":0, "Contacted - Doesn't Qualify":0, "Unable to contact - Bad Contact Information":0, "EP Confirmed":0, "Contacted - Invalid Profile":0, "EP Sent":0, "EP Received":0})
```

```python
# % of leads that were evaluated
df['LeadClosedFlag'].sum()/len(df) * 100
```

```python
df['LeadQuality'] = df['LeadCategory'].map({'Good':1, 'Bad':0, 'Unknown':0})
```

```python
df
```

```python
df['LeadCreated'].min()
```

```python
df['LeadCreated'].max()
```

```python
# start date of the week when the lead_date was created
df['Week'] = df['LeadCreated'].dt.to_period('W').apply(lambda r: r.start_time)
```

```python
df['Week']
```

```python
weekly = df.groupby(['Week']).agg(total_leads=('LeadQuality', 'count'), good_rate=('LeadQuality', 'mean'), closed_rate=('LeadClosedFlag', 'mean'), unknown_rate=('LeadCategory', lambda x: (x=='Unknown').mean())).reset_index()
```

```python
weekly
```

```python
plt.figure(figsize=(10,5))
plt.plot(weekly['Week'], weekly['good_rate'])
plt.xticks(rotation=45)
plt.xlabel("Week")
plt.ylabel("Good Rate")
plt.title("Weekly Lead Quality Trend")
plt.show()
```

```python
weekly['week_index'] = range(len(weekly))
slope, intercept, r_value, p_value, std_err = linregress(
    weekly['week_index'],
    weekly['good_rate']
)

print("Slope:", slope)
print("P-value:", p_value)
```

```python
plt.figure(figsize=(10,5))
plt.plot(weekly['Week'], weekly['closed_rate'])
plt.xticks(rotation=45)
plt.xlabel("Week")
plt.ylabel("Closed Rate")
plt.title("Weekly Closed Rate Trend")
plt.show()
```

```python
weekly['week_index'] = range(len(weekly))
slope, intercept, r_value, p_value, std_err = linregress(
    weekly['week_index'],
    weekly['closed_rate']
)

print("Slope:", slope)
print("P-value:", p_value)
```

```python
plt.figure(figsize=(10,5))
plt.plot(weekly['Week'], weekly['total_leads'])
plt.xlabel("Week")
plt.ylabel("Total Leads")
plt.title("Weekly Lead Volume")
plt.xticks(rotation=45)
plt.show()
```

```python
weekly['week_index'] = range(len(weekly))
slope, intercept, r_value, p_value, std_err = linregress(
    weekly['week_index'],
    weekly['total_leads']
)

print("Slope:", slope)
print("P-value:", p_value)
```

```python
# Q1 Conclusions: -

# Weekly Closed Rate Trend
# -> Closed Rate is declining over time
# -> The decline is statistically significant.
# -> This is likely a real deterioration, not just random fluctuation.

# Conclusion: Lead-to-customer conversion is worsening over time.

# Weekly Good Rate Trend
# -> Good rate is basically flat.
# -> Any movement is statistically insignificant.
# -> No clear upward or downward pattern

# Conclusion: Lead quality is stable over time.

# Weekly Volume Trend
# -> Lead volume is declining over time.
# -> The drop is statistically significant.

# Conclusion: Traffic volume is shrinking steadily.
```

```python
df['SearchQuery'].value_counts().reset_index().head()
```

```python
def traffic_type(x):
    if pd.isna(x['SearchQuery']) and x['Partner'] == 'call_center':
        return 'Call Center'
    elif pd.isna(x['SearchQuery']) and x['Partner'] != 'call_center':
        return 'Display'
    else:
        return 'Search'
```

```python
df['TrafficType'] = df.apply(traffic_type, axis=1)
```

```python
# df.groupby(['Partner'])[['SearchQuery']].apply(lambda x: x.notna().mean()).reset_index()
```

```python
trafficSummary = df.groupby(['TrafficType']).agg(total_leads=('VendorLeadID', 'count'), good_rate=('LeadQuality', 'mean'), closed_rate=('LeadClosedFlag', 'mean')).reset_index()
trafficSummary
```

```python
# Conclusion:-
# Search and Call Center traffic generate higher Good and Closed rates compared to Display. 
# Display traffic, while highest in volume, underperforms in quality and is likely diluting overall performance.
```

```python
# Q3: Can We Increase Closed Rate from 8% to 9.6%?

## Is It Achievable?

# Yes, Several segments already have closed rates above 9.6%:

# ->Google Search (~11%)
# -> Debt Holding Tank (~14–15%)
# -> Financial Services (~12%)
# -> 70–90k debt segment (~13%)

# This shows the target is structurally achievable.


# Where Do We See Opportunities?

## 1. Reallocate Budget Away from Google Display

# Google Display has ~5% closed rate but drives large volume.
# Reducing exposure here and reallocating toward higher-performing Search traffic (~11%) would improve overall weighted average closed rate.
# Large volume + low performance = biggest drag. Even small mix shift moves average meaningfully.


## 2. Increase Focus on Mid-Range Debt Segments

# Very low debt (7.5–10k) closes at ~5%, while mid-range bands (10–15k, 70–90k) close at ~13–14%.
# Refining targeting to prioritize mid-range debt bands can lift overall quality.
# Strong difference in conversion rate across debt levels.


## 3. Scale High-Performing Campaigns

# Campaigns like:

# -> Debt Holding Tank (~14.5%)
# -> Financial Services (~12%)

# Outperform large generic campaigns like:

# -> DebtReductionInc (~6%)

# Shifting budget toward better-performing campaigns should improve results.
# Campaign performance gap is material and actionable.


## Overall View

# The 9.6% target does not require improving every segment. It can likely be achieved by:

# -> Reducing exposure to lowest-performing traffic
# -> Scaling better-performing segments
# -> Refining customer targeting

# At current volume (~3,000 leads), an 8% closed rate implies ~240 closed deals.
# To reach 9.6%, we would need ~288 closures — roughly 48 additional deals.
# Given the performance gap between low and high performing segments (e.g., 5% vs 11–14%), modest traffic reallocation could realistically bridge this gap.
```

```python
# Q3: Can We Increase Lead Quality by 20%?

# -> Current Good Rate ≈ 13%
# -> Target (20% lift) ≈ 15–16%

## Is It Achievable?

# Yes, There are already segments delivering Good Rates above 15%:

# -> Adknowledge (~18–21%)
# -> Debt Holding Tank (~18%)
# -> Financial Services (~18–19%)
# -> 70–90k Debt Segment (~19%)
# -> Google Search (~14–15%)

# Since parts of the traffic mix already perform at or above the target level, the goal is achievable through optimization rather than structural change.


# Where Are the Opportunities?

## 1. Reduce Low-Quality Display Traffic

# -> Google Display drives high volume but has a lower Good Rate (~9%).
# -> Because of its scale, it drags down the overall average.
# -> Shifting even a portion of this traffic toward stronger-performing Search channels can improve overall lead quality.


## 2. Focus More on Mid-Range Debt Profiles

# -> Debt ranges such as 10–15k and 70–90k produce Good Rates around 18–19%.
# -> Lower debt segments (7.5–10k) perform much worse (~7%).
# -> Refining targeting toward higher-performing debt profiles can lift weighted average quality without increasing total volume.



## 3. Allocate More Budget to Stronger Campaigns

# Some campaigns materially outperform broad ones:

# -> Debt Holding Tank (~18%)
# -> Financial Services (~19%)

# Compared to:

# -> DebtReductionInc (~11%)
# -> Debt General (~8–9%)

# Shifting allocation toward better-performing campaign structures can raise overall quality.


# Overall Assessment

# Improving lead quality by 20% does not require every segment to improve.

# It can likely be achieved through:

# -> Reducing reliance on lower-quality Display traffic
# -> Increasing share of higher-performing Search traffic
# -> Targeting stronger debt segments
# -> Scaling structurally better campaigns
```
